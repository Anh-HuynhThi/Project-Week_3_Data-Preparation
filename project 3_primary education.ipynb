```python
import pandas as pd
import numpy as np 
```

```python
data=pd.read_excel(r'C:\Users\James\Documents\CODE DATA ANALYSIS\Module 1\Project 3\Primary Education.xlsx')
data.head()
```

```python
data.info()
```

```python
#FIND AND REPLACE THE MISSING VALUE
data.isna().sum().sort_values(ascending=False)
```

```python
#FIND MISING AND REPLACE 'REGION':
data[data['Region'].isna()]
```

```python
data['Region']=data['Region'].replace(np.nan,'ECA')
```

```python
#FIND MISSING AND REPLACE 'DATA SOURCE':
data[data['Data source'].isna()]
```

```python
list(data['Data source'].mode())
```

```python
data['Data source']=data['Data source'].replace(np.nan,'Multiple Indicator Cluster Survey')
```

```python
#FIND MISING AND REPLACE'SUB-REGION':
data[data['Sub-region'].isna()]
```

```python
data['Sub-region']=data['Sub-region'].replace(np.nan,'MENA')
data['Sub-region']=data['Sub-region'].replace(np.nan,'EECA')
data['Sub-region']=data['Sub-region'].replace(np.nan,'EECA')
data['Sub-region']=data['Sub-region'].replace(np.nan,'ESA')
data['Sub-region']=data['Sub-region'].replace(np.nan,'EECA')
```

```python
#FINDING AND REPLACE 'Income Group':
data[data['Income Group'].isna()]
```

```python
data['Income Group']=data['Income Group'].replace(np.nan,'Lower Middle Income(LM)')
data['Income Group']=data['Income Group'].replace(np.nan,'Upper Middle Income(UM)')
data['Income Group']=data['Income Group'].replace(np.nan,'Upper Middle Income(UM)')
data['Income Group']=data['Income Group'].replace(np.nan,'Low Income(L)')
data['Income Group']=data['Income Group'].replace(np.nan,'Lower Middle Income(LM)')
data['Income Group']=data['Income Group'].replace(np.nan,'Lower Middle Income')
```

```python
#REPLACE MISING AND REPLACE 'TOTAL':
data[data['Total'].isna()]     
```

```python
data['Total']=data['Total'].replace(np.nan,data['Total'].mean())
```

```python
#FINDING MISSING AND REPLACE RESIDENCE URBAN:
data[data['Residence Urban'].isna()]
```

```python
data['Residence Urban']=data['Residence Urban'].replace(np.nan,data['Residence Urban'].mean())
```

```python
#FINDING MISSING AND REPLACE 'RESIDENCE RURAL':
data[data['Residence Rural'].isna()]
```

```python
data['Residence Rural']=data['Residence Rural'].replace(np.nan,data['Residence Rural'].mean())
```

```python
#FINDING MISSING AND REPLACE 'Wealth quintile Poorest' 
data[data['Wealth quintile Poorest'].isna()]
```

```python
data['Wealth quintile Poorest']=data['Wealth quintile Poorest'].replace(np.nan,data['Wealth quintile Poorest'].mean())
```

```python
#FINDING MISSING AND REPLACE 'Wealth quintile Richest ' 
data[data['Wealth quintile Richest'].isna()]
```

```python
data['Wealth quintile Richest']=data['Wealth quintile Richest'].replace(np.nan,data['Wealth quintile Richest'].mean())
```

```python
# TEST MISSING VALUE
data.isna().sum().sort_values(ascending=False)
```
